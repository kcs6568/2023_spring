```python
import numpy as np
import matplotlib.pyplot as plt

task = "cifar10"
root = "/root/volume/vis_result/"
filename = f"{task}_epoch.png"
save_path = root + filename

baseline = [86.46, 89.4, 91.25, 92.12, 91.86, 91.74, 93.57, 93.75, 95.73, 95.67, 95.79, 95.94]
pcgrad = [69.32, 76.88, 83.19, 85.72, 87.36, 89.08, 90.27, 91.53, 92.31, 92.99, 93.56, 93.82]
gradvac = [68.16, 75.13, 83.26, 85.21, 86.94, 88.73, 90.31, 91.28, 91.92, 93.14, 93.48, 93.75]
cagrad = [80.55, 83.82, 87.28, 88.5, 89.97, 91.45, 91.4, 92.05, 93.94, 94.14, 94.12, 94.7]

x_len = len(baseline)
x_label = [str(i+1) for i in range(x_len)]

plt.plot(baseline, '-', color='black', marker='o', markersize=8, label='Baseline')
plt.plot(pcgrad, '-.', color='darkblue', marker='>', markersize=8, label='PCGrad')
plt.plot(gradvac, ':', color='green', marker='d', markersize=7, label='GradVac')
plt.plot(cagrad, '-', color='red', marker='s', markersize=7, label='CAGrad')

# plt.plot(t1, '--', color='darkblue', marker='x', markersize=8, label='CIFAR10')
# plt.plot(t2, '-.', color='cornflowerblue', marker='X', markersize=9, label='MINICOCO')
# plt.plot(t3, '-', color='darkblue', marker='o', markersize=8, label='VOC2012')

# plt.plot(
#     zero_to_all,
#     color="brown",
#     linestyle="--")
plt.grid()
# plt.fill_between(x_label[0:4], zero_to_all[0:4], alpha=0.5)
# plt.fill_between(x_label[3:], zero_to_all[3:], alpha=0.5, color='Coral')

plt.xticks(np.arange(0, x_len), labels=x_label)
plt.yticks(np.arange(65, 100, 5))
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Static Approach for CIFAR10")
# plt.legend(loc=7, fontsize=15)
plt.legend()
plt.savefig(save_path)
plt.show()
```
